```python
#list comprehension masterclass
scores=[45,82,91,33,67,55,78,29,95,61]
names=["rahul","priya","arjun","sneha","vikram"]
a=[a for a in scores if a>=60]
print(a)
print("______________")

d=["a" if  d>=80 else "b" if d>=60 else "c" for d in scores]
print(d)
print("___________________")
e=[e**2 for e in scores]
print(e)
print("______________________")
f=[f for f in scores  if f%2==0]
print(f)
print("_________________")
m=[m for m in names]
print(m)

```

```python
#fizzbuzz-the classic
def fizzbuzz(n):
    b=["fizzbuss"if i%15==0 else "fizz" if i%3==0 else "buzz" if i%5==0 else str(i) for i in range(1,n+1)]
    print(b)
fizzbuzz(20)
```

```python
#number to words
ones=["","one","two","three","four","five",
        "six","seven","eight","nine","ten",
      "eleven","twelve","thirteen","fourteen","fifteen",
      "sixteen","seventeen","eighteen","nineteen"]
tens=["","","twenty","thirty","forty","fifty",
     "sisty","seventy","eighty","ninety"]
def to_words(n):
    if n==0:
        return "Zero"
    if n<20:
        return ones[n]
    if n<100:
        return tens[n//10] + " "+ones[n%10]
    if n<1000:
        return ones[n//100] + "hundred" + to_words(n%100)
for num in [0,7,15,42,100,305,999]:
    print(num,to_words(num))
```
