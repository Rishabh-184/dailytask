```python
#common friends finder
rahul_friends = {"Amit", "Priya", "Rohan", "Sneha", "Karan"}
priya_friends = {"Rohan", "Sneha", "Ankit", "Divya"}
for i in rahul_friends:
    for j in priya_friends:
        if i==j:
            print(i)
print("______________")
    
for i in rahul_friends:
    if i not in priya_friends:
        print(i)
print("______________")

a=set()
for i in rahul_friends:
    for j in priya_friends:
        a.add(i)
        a.add(j)
print(a)
print(len(a))
print("______________")
    
```
