```python
#number pattern 
def print_triangle(n):
    for i in range(1,n+1):
        for j in range(1,i+1):
            print("*",end=" ")
        print()
def print_square(n):
    for i in range(n):
        for j in range(n):
            print("*",end=" ")
        print()
def print_reverse(n):
    for i in range(1,n+1):
        for j in range(1,7-i):
            print("*",end=" ")
        print()


        
print_triangle(5)
print()
print_square(4)
print()
print_reverse(5)
```
