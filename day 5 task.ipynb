```python
#staircase
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n* factorial(n-1)

def ways(n):
    if n==0 or n==1:
        return 1
    else:
        return ways(n-1)+ways(n-2)

print(factorial(5))
print(factorial(0))
print(ways(1))
print(ways(3))
print(ways(5))
```
