```python
def is_armstrong(n):
    t=n
    n=str(n)
    if len(n)>1:
        l=len(n)
        ar=0
        while n>0:
            r=n%10
            ar=ar+r**l
            n=n//10
        if t==ar:
            return True
    else:
        return False

def is_prime(n):
    if n==0 or n==1:
        return False
    else:
        for i in range(2,n):
            if n%i==0:
                return False
        return True
def classify(n):
    lables=[]
    if is_armstrong: lables.append("armstrong")
    if is_prime(n): lables.append("prime")
    return lables if lables else ["None"]
for n in [2,11,121,153,370,1634]:
    print(n,classify(n))
```
